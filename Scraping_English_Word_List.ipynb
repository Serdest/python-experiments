Goals:

*   Scrape Complete English Word List from http://www.yougowords.com/english-word-list
*   Write word list to `.csv` file



```
import pandas as pd
```

```
base_url = "http://www.yougowords.com/english-word-list-"

url_list = [base_url + str(i) for i in range(1,1981)]
```

```
combined = pd.DataFrame()

# change "10" to "len(url_list)" for scraping all pages

for i in range(10):
  combined = combined.append(pd.read_html(url_list[i])[0].drop(['Favourite'],axis=1))
```

```
combined.to_csv("combined.csv",index=False)
pd.read_csv("combined.csv")
```
